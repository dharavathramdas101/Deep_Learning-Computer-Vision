```python
import numpy as np
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow
```

```python
img1 = cv2.imread(r"/content/traffic_sig.png")
```

```python
cv2_imshow(img1)
```

```python
img2 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
cv2_imshow(img2)
```

```python
img3=cv2.GaussianBlur(img1,(7,7),cv2.BORDER_DEFAULT)
cv2_imshow(img3)
```

```python
blur_img = cv2.GaussianBlur(img1,(11,11),cv2.BORDER_DEFAULT)
cv2_imshow(blur_img)
```

```python
b_img = cv2.GaussianBlur(img1,(5,5),cv2.BORDER_CONSTANT)
cv2_imshow(b_img)
```

```python
img_c=cv2.Canny(b_img,125,175)
cv2_imshow(img_c)
```

```python
d_img = cv2.dilate(img_c,(3,3),iterations=3)
cv2_imshow(d_img)
```

```python
d1_img = cv2.dilate(img_c,(11,11),iterations=3)
cv2_imshow(d1_img)
```
