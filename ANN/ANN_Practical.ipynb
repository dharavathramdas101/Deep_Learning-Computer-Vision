```python
X=df.drop("class",axis=1)
```

```python
y=df['class']
```

```python
## categorical to numerical columns using encoding
```

```python
y=pd.get_dummies(y,drop_first=True)
```

```python
y.head()
```

```python
## tain test split split 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## standard scaler required because of optimizer used to convert local minima
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

X_train=sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# ANN
# built tensorflow model goto tensorflow document python tensorflow keras
## numpy array 1d,2d,3d
# tensorflow inside we have keras we import keras
# sequential 
#input layer after hidenlayer after output layer 
## input 170 after unit 10 means layer in hiden layer1  
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()

classifier.add(Dense(units=10,kernel_initializer='he_uniform', activation="relu",input_dim=170))

classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation="relu"))

classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation="sigmoid"))
```

```python
# farword and backward with the help of  sequential model
# optimizer to do backpropagation
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
### summary of classifier
## 170 input*10 weight+10*1 bias  == 1710
### 10 * 10 weight + 10*1 bias  == 110
## 10 * 1 weight + 1*1 bias  == 11
```

```python
classifier.summary()
```

```python
## train model 
## batch size default 32 so we 10
## epochs = 30 
## validation split is 25%
## data size 4000 
## epochs = 20, batch size = 10 this input layer
## 1epoch = fp +bp
## batch size =10  == 4000/10 == 400  in one batch we provide 400 rows
## total 400 * 10 batches means 10 itertations
```

```python
model_history = classifier.fit(X_train,y_train,batch_size=10,epochs=30,validation_split=0.25)
```

```python
## hyperparameter tuning
```

```python
from keras_tuner.tuners import RandomSearch
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units=32,kernel_initializer = 'he_uniform',activation='relu',input_dim=170))
  model.add(Dense(units = 32, kernel_initializer = 'he_uniform',activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop","adadelta"])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
## i want to check val accuracy 
```

```python
tuner=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project1',project_name='aps-fault')
     
```

```python
## give best val accuracy
```

```python
tuner.search(X_train,y_train,validation_data=(X_test,y_test),epochs=5)
```

```python
##give best optiizer 
```

```python
from tensorflow.keras import optimizers
def build_model(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=32),
                               activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
tuner_second=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project3',project_name='aps-fault-sensor')
```

```python
tuner_second.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
