```python
import cv2
import numpy as np
import pandas as pd
import pytesseract
from matplotlib import pyplot as plt
from pytesseract import Output
import re
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread(r"/content/ts_numberplate_img.jpg")
```

```python
gray_img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
```

```python
remove_noise = cv2.medianBlur(gray_img,5)
```

```python
cv2_imshow(remove_noise)
```

### canny edge detection

```python
edge_img = cv2.Canny(remove_noise,150,200)
```

```python
cv2_imshow(edge_img)
```

```python
custom_config = r'--oem 3 --psm 6'
```

```python
pytesseract.image_to_string(img, config=custom_config)
```
